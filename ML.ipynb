```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Step 1: Load the dataset
df = pd.read_csv('dataset_MLproject.csv')  # Replace with your CSV file path

# Step 2: Define the feature and create the target based on the condition
# Assuming the column you're interested in is named 'FeatureColumn'
df['Target'] = (df['Edges integer'] > 50).astype(int)

# Step 3: Separate the features and target
X = df[['Edges integer']].values  # Replace 'FeatureColumn' with your column name
y = df['Target'].values

# Step 4: Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Step 5: Standardize the features (optional but recommended)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Step 6: Train Logistic Regression model
model = LogisticRegression()
model.fit(X_train_scaled, y_train)

# Step 7: Make predictions
y_pred = model.predict(X_test_scaled)

# Step 8: Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Plot Confusion Matrix
fig, ax = plt.subplots(figsize=(5, 5))
plot_confusion_matrix(model, X_test_scaled, y_test, ax=ax)
plt.title('Confusion Matrix')
plt.show()

# Step 9: 2D Sigmoid Plot
xx = np.linspace(-3, 3, 100)
yy = model.predict_proba(xx.reshape(-1, 1))

plt.plot(xx, yy[:, 1], label='Sigmoid Curve', color='blue')
plt.xlabel('Feature Value')
plt.ylabel('Probability')
plt.title('2D Sigmoid Plot')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('dataset_MLproject.csv')
df.head()
```

```python
df['Target'] = (df['Edges integer'] > 50).astype(int)
```

```python
X = df[['Edges integer']].values  # Replace 'FeatureColumn' with your column name
y = df['Target'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
xx = np.linspace(-3, 3, 100)
yy = model.predict_proba(xx.reshape(-1, 1))
# Print all points (feature values and corresponding probability for class 1)
for feature, prob in zip(xx, yy[:, 1]):
    print(f"Feature: {feature:.2f}, Probability of class 1: {prob:.4f}")

```

```python
plt.plot(xx, yy[:, 1], label='Sigmoid Curve', color='blue')
plt.xlabel('Feature Value')
plt.ylabel('Probability')
plt.title('2D Sigmoid Plot')
plt.grid(True)
plt.show()
```
